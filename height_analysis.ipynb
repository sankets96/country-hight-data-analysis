```python
#male has max height
max_h1 = df['Male Height in Cm'].max()
max_h1
```

```python
#find the wich country has max hight
max_hi_m = df['Male Height in Cm'].idxmax()
max_m_country = df['Country Name'][max_hi_m]
max_m_country
```

```python
#which country female has max height
max_h2 = df['Female Height in Cm'].max()
max_h2
```

```python
#find the index
max_high_f = df['Female Height in Cm'].idxmax()
max_f_country = df['Country Name'][max_high_f]
max_f_country
```

```python
max = df.sort_values('Male Height in Cm',ascending=False).head(10)
max
```

```python
#top 10
plt.figure(figsize=(10,7))
sns.barplot(data=max,x='Country Name',y='Male Height in Cm')
```

```python
#top 10
plt.figure(figsize=(10,7))
sns.barplot(data=max,x='Country Name',y='Female Height in Cm')
```

```python
#male has minium height
min_h1 = df['Male Height in Cm'].min()
min_h1
```

```python
#which country male has minium height

min_hi_m = df['Male Height in Cm'].idxmin()
min_m_country = df['Country Name'][min_hi_m]
min_m_country
```

```python
#female has minium height
min_h2 = df['Male Height in Cm'].min()
min_h2
```

```python
min_hi_f = df['Male Height in Cm'].idxmin()
min_f_country = df['Country Name'][min_hi_f]
min_f_country
```

```python
a = [max_h1,max_h2]  #max hight
b = [max_m_country,max_f_country]  #max hight country
c= [min_h1,min_h2]  #min hight
d=[min_m_country,min_f_country]  #min hight country

```

```python
pd.DataFrame({'Max hight' :a,'max hight country' :b,'min hight' :c,'min hight country':d},index=['MALE','FEMALE'])
```
